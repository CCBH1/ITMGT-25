```python
import requests
import matplotlib.pyplot as plt
import time as t
import numpy as np
import pandas as pd

%matplotlib inline

fig, ax = plt.subplots(figsize=(10, 6))
x = []
y = []

i = 1
while(i <= 100):
    
    r = requests.get(url = "http://api.open-notify.org/iss-now.json")
    position = r.json()

    location = position["iss_position"]
    lat = location["latitude"]
    lon = location["longitude"]
    time = position["timestamp"]
    
    x.append(lat)
    y.append(lon)
    
    plt.scatter(x,y)
    plt.title('Path of the ISS')
    plt.xlabel('Latitude')
    plt.ylabel('Longitude')
    plt.grid()
    
    result = str(i) + ". Latitude: " + str(lat) + "; Longitude: " +str(lon) + "; Timestamp: " +str(time)
    print(result)
    t.sleep(10)
    i = i + 1
```
